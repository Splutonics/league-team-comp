```python
import random
all_champs = json.load(open('data/all_champs.json', 'r'))
```

```python
random.seed(44)
champs_1 = random.choices(all_champs,k=10)
champs_2 = random.choices(all_champs,k=10)
champs_3 = random.choices(all_champs,k=10)
champs_4 = random.choices(all_champs,k=10)
champs_5 = random.choices(all_champs,k=10)
#
# del all_champs
```

```python
# https://stackoverflow.com/questions/19744542/itertools-product-eliminating-repeated-elements
# gives all unique combinations of multiple lists
def uprod(*seqs):
    def inner(i):
        if i == n:
            yield tuple(result)
            return
        for elt in sets[i] - seen:
            seen.add(elt)
            result[i] = elt
            for t in inner(i+1):
                yield t
            seen.remove(elt)

    sets = [set(seq) for seq in seqs]
    n = len(sets)
    seen = set()
    result = [None] * n
    for t in inner(0):
        yield t

pared_champs_combos = uprod(champs_1,champs_2,champs_3,champs_4,champs_5)
# pared_champs_combos = list(uprod(champs_1,champs_2,champs_3,champs_4,champs_5))

# TODO should be a factor 
# len(pared_champs_combos)
```

```python
all_combo_scores = pd.DataFrame()
all_combo_scores['champs'] = None
all_combo_scores['scores'] = None
all_combo_scores['total_score'] = None
idx = 0
for combo in pared_champs_combos:

    # ? currently ~80-90 seconds per 10000 combos
    # //
    if idx == 100:
        break

    scores = list()
    for champ1, champ2 in permutations(combo,2):
        try:
            synergy_score = synergy_df.loc[(synergy_df.champion == champ1) & (synergy_df.other_champion == champ2)]['synergy'].iloc[0]
            scores.append(synergy_score)
        except:
            continue
    # print(scores)
    idx += 1

    all_combo_scores = all_combo_scores.append({'champs':combo,'scores':scores,'total_score':sum(scores)},ignore_index=True)


all_combo_scores.sort_values('total_score',inplace=True,ascending=False)


all_combo_scores


        
```
